```python
# import modules
from datetime import datetime
import yfinance as yf
import matplotlib.pyplot as plt

# initialize parameters
start_date = datetime(2025, 1, 1)
end_date = datetime(2025, 9, 1)


start_date = '2025-01-01'
end_date = '2025-09-01'

# get the data
data = yf.download('AAPL', start = start_date,
                   end = end_date)

# display
plt.figure(figsize = (20,10))
plt.title('Opening Prices from {} to {}'.format(start_date,
                                                end_date))
plt.plot(data['Open'], label='Open')
plt.plot(data['Close'], label='Close')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
import yfinance as yf
import contextlib
import sys
import io

def is_valid_ticker(ticker: str) -> bool:
    f = io.StringIO()
    with contextlib.redirect_stderr(f):  # capture yfinance error logs
        try:
            info = yf.Ticker(ticker).get_info()
            return bool(info and "symbol" in info)
        except Exception:
            return False

print(is_valid_ticker('roro'))
```
